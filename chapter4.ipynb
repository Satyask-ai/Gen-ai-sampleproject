```python
from transformers import GPT2Tokenizer, GPT2Model
import  torch
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tokenizer= GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2Model.from_pretrained("gpt2", output_attentions=True)
```

```python
text = "The AI is smart"
inputs = tokenizer (text, return_tensors="pt")
with torch.no_grad():
    outputs = model(**inputs)
```

```python
attentions = outputs.attentions
last_layer_attention = attentions[-1]
```

```python
attention_matrix = last_layer_attention.squeeze(0).mean(0).numpy()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(attention_matrix, xticklabels=inputs['input_ids'][0].tolist(),
            yticklabels=inputs['input_ids'][0].tolist(), cmap="viridis", annot=True, fmt=".2f")
tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])
plt.xticks(ticks=range(len(tokens)), labels=tokens, rotation=90)
plt.yticks(ticks=range(len(tokens)), labels=tokens )
plt.title("Self-Attention Heatmap (GPT-2, Last Layer)")
plt.tight_layout()
plt.show()
```
